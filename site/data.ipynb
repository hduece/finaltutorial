# Part 1: Analysis of COVID Vaccines at the State Level

Part 1 Description: The following DataFrame contains information regarding the number of people vaccinated in each state in the United States. It contains information about the distrbution of the three different available vaccines (Moderna, Pfizer, and Janssen) and the number of doses people in each state have received. There is also data regarding vaccine rates for the 65+ age group, since this age group is a high-risk group for the virus. This data can be compared to the socio-economic patterns in each state to see if any of these factors are correlated with vaccine rates.

```python
#Set up pandas and import our CSV files to DataFrames:
import pandas as pd
pd.DataFrame
vac_state = pd.read_csv("./data/vaccine_state.csv")
vac_county = pd.read_csv("./data/Vaccine_county.csv")
stats = pd.read_csv("./data/countystats.csv")
trans = pd.read_csv("./data/county_transmission.csv")
gdp = pd.read_csv("./data/county_gdp.csv")
state_income_df = pd.read_csv("./data/StateIncome.csv")

#Get rid of max rows so whole dataframe can be examined, fix a few things about our data frame to make our lives easier:
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
vac_state = vac_state.set_index("State")
vac_state = vac_state.rename(index = {'New York State': "New York"})
vac_state = vac_state.drop(['United States'], axis = 0)
vac_state
```

```python
# All of the data types are numerical
vac_state.dtypes
```

# Obtaining State Income Data
This section explores how the median household income of each state correlates to what percent of each state's population has received at least one dose of any COVID vaccine.

```python
#Set up the state income DataFrame (already defined in the imports section):
state_income_df = state_income_df.set_index("State")
state_income_df = state_income_df.rename(columns = {"HouseholdIncome" : "Household_Income"})
state_income_df
```

```python
#Combine the vaccination by state dataframe with this state income dataframe:
vaccine_income_df = state_income_df.join(vac_state, how = "outer")
vac_state = vac_state.drop(['District of Columbia'], axis = 0)
vaccine_income_df
# vaccine_income_df.dtypes 
# all dtypes are numerical
```

```python
# Create scatterplot to visualize relationship between income and percent of population that has been vaccinated
import matplotlib.pyplot as plt
vaccine_income_df.plot(kind = "scatter", x = 'Household_Income', y = 'Percent of Total Pop with at least One Dose by State of Residence', color = 'red', title = "Income vs. Percent Population Vaccinated")
plt.show()
```

```python
# calculate the correlation between household income and percent of populations vaccinated
variables = ["Household_Income", 'Percent of Total Pop with at least One Dose by State of Residence']
vaccine_income_df[variables].corr()
```

### Household Income Analysis:
The correlation between income and having at least one dose of the vaccine is 0.630108, which is decently high. There is a moderate correlation between these two variables. 

# Obtaining State Demographics Data
The state demographics data set contains state-level data regarding education levels, ethinicity and race distributions, employment rates, and other socio-economic factors. Each of these factors can be analyzed individually in relation to vaccine rates in each state. 

```python
state_demo_df = pd.read_csv("data/state_demographics.csv")
state_demo_df = state_demo_df.set_index("State")
state_demo_df = state_demo_df.drop(['District of Columbia'], axis = 0)
state_demo_df
# state_demo_df.dtypes
# all dtypes are numerical
```

```python
# create new data frame containing demographics and vaccine data 
vaccine_education_df = state_demo_df.join(vac_state, how = "outer")
vaccine_education_df
# vaccine_education_df
# all dtypes are numerical
```

# High School Education vs. Vaccination Percentage:

```python
# create scatterplot to visualize relationship between high school education and percent of population that has been vaccinated
vaccine_education_df.plot(kind = "scatter", x = 'Education.High School or Higher', y = 'Percent of Total Pop with at least One Dose by State of Residence', color = 'red', title = "High School Education vs. Percent Population Vaccinated")
plt.show()
# calculate correlation for this relationship
hs_variables = ["Education.High School or Higher", 'Percent of Total Pop with at least One Dose by State of Residence']
vaccine_education_df[hs_variables].corr()
```

### High School Education Analysis
There is not a very strong correlation between high school education and the percent of the total population that has receivied at least one dose of a vaccine. The correlation is 0.147417, and the scatter plot shows only a slight positive correlation between these variables

# College Education vs. Vaccination Percentage:

```python
# create scatterplot to visualize relationship between college education and percent of population that has been vaccinated
vaccine_education_df.plot(kind = "scatter", x = "Education.Bachelor's Degree or Higher", y = 'Percent of Total Pop with at least One Dose by State of Residence', color = 'red', title = "College Education vs. Percent Population Vaccinated")
plt.show()
# calculate correlation for this relationship
college_variables = ["Education.Bachelor's Degree or Higher", 'Percent of Total Pop with at least One Dose by State of Residence']
vaccine_education_df[college_variables].corr()
```

### College Education Analysis
There is a strong positive correlation between having a bachelor's degree or higher and the percent of the total population that has received at least one dose of a vaccine. The correlation is 0.729796, and the scatter plot shows only a strong positive correlation between these variables. This indicates that states with more educated populations may be more likely to have a larger percent of their poopulation vaccinated.

# Population Density vs. Vaccination Percentage:

```python
# Create scatterplot to visualize relationship between population density of a state and percent of population that has been vaccinated
vaccine_education_df.plot(kind = "scatter", x = 'Population.Population per Square Mile', y = 'Percent of Total Pop with at least One Dose by State of Residence', color = 'red', title = "Population Density vs. Percent Population Vaccinated")
plt.show()
# calculate correlation for this relationship
density_variables = ["Population.Population per Square Mile", 'Percent of Total Pop with at least One Dose by State of Residence']
vaccine_education_df[density_variables].corr()
```

# Under 18 Percentage vs Vaccine Percentage

```python
# Create scatterplot to visualize relationship between population density of a state and percent of population that has been vaccinated
vaccine_education_df.plot(kind = "scatter", x = 'Age.Percent Under 18 Years', y = 'Percent of Total Pop with at least One Dose by State of Residence', color = 'red', title = "Population Density vs. Percent Population Vaccinated")
plt.show()
# calculate correlation for this relationship
density_variables = ["Age.Percent Under 18 Years", 'Percent of Total Pop with at least One Dose by State of Residence']
vaccine_education_df[density_variables].corr()
```

This could be due to older populations having a higher risk associated with COVID-19, which is a factor that can convince older people to get the vaccine according to the Health Belief model in public health. Another factor in this may be that older populations had access to the vaccine much earlier than younger populations.
